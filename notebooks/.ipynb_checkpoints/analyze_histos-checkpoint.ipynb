```python
# Load Data
FIT_PAR_IDXS = {"spe": (3,5), "lyso": (3,5), "sodium": (1,2), "cesium": (1,2), "cobalt": (1,2)}


modules, module_serial_nums = {}, []
for fname in MODULE_PATHS:
    if "root" not in fname:
        continue

    stats = fname.split("_")s
    if stats[1] not in module_serial_nums:
        module_serial_nums.append(stats[1])

    source = [s for s in SOURCES if s in stats[4]][0]
    module_id =  f"{stats[1]}_{source}"

    modules[module_id] = {
        "ov" : float(stats[2][3:]),
        "source" : source,
        "n_spe" : int(stats[3][4:]),
        "n_source" : int(stats[4][len(source)+1:-5]),
        "tfile" : TFile(F"{DATA_DIR}/{fname}"),
        # "tfile_keys": [k.GetName() for k in TFile(F"{DATA_DIR}/{fname}").GetListOfKeys()],
        "spe_charge": np.empty((32,2), dtype=float), # channels x (val, err)
        "source_charge": np.empty((32,2), dtype=float), # channels x (val, err)
    }
    modules[module_id]["tfile_keys"] = [k.GetName() for k in modules[module_id]["tfile"].GetListOfKeys()]

    ## Extract Channel SPE & Source Charges
    for s in ("spe", "source"):
        _s = source if s == "source" else s
        iAve, iErr = FIT_PAR_IDXS[_s]

        for ch in CHANNELS:
            k = f"{_s}_ch{ch}_fit"
            if k not in modules[module_id]["tfile_keys"]:
                continue

            tfit = modules[module_id]["tfile"].Get(k)
            if iAve == iErr:
                modules[module_id][f"{s}_charge"][ch] = [tfit.GetParameter(iAve), tfit.GetParError(iAve)]
            else:
                modules[module_id][f"{s}_charge"][ch] = [tfit.GetParameter(iAve), tfit.GetParameter(iErr)]

module_sources = ["spe"] + [ s for s in SOURCES if s in [m["source"] for mid,m in modules.items()] ]
```

```python
# Plot SPE for each bar

## Settings
serial_num = module_serial_nums[0]
source = ("spe", "lyso", "sodium", "cesium", "cobalt")[2]
module_id = f"{serial_num}_{'lyso' if source == 'spe' else source}"

print(f"{serial_num=}")
print(f"{module_id=}")
print(f"{source=}")

## Plots
n_rows, n_cols = max(floor(np.sqrt(len(BARS))),3), max(ceil(np.sqrt(len(BARS))),3)
canvas = TCanvas("","",TPAD_WIDTH*n_cols,TPAD_HEIGHT*n_rows)
canvas.Divide(n_cols, n_rows)
tlegend_coords = (0.8, 0.75, 0.97, 0.87)

module = modules[module_id]
tfile = module["tfile"]

### Create Histos
hists, legends, hmin, hmax = [], [], 1e99, -1
for bar in BARS:
    chs = [bar, bar+16] # [left_ch, right_ch]
    nns = [f"Ch {ch}" for ch in chs]
    hhs = [tfile.Get(f"{source}_ch{ch}") for ch in chs]
    tts = [f"Bar {bar} {source.upper()} Response"] * len(chs)
    lws = [4] * len(chs)
    lcs = [rt.kAzure, rt.kOrange]

    legend = TLegend(*tlegend_coords)
    for ihh, hh in enumerate(hhs):
        hh.SetTitle(tts[ihh])
        hh.SetName(nns[ihh])
        hh.SetLineColor(lcs[ihh])
        hh.SetLineWidth(lws[ihh])
        legend.AddEntry(hh, f"{nns[ihh]}", "l")
        hmin = min(hmin, hh.GetBinContent(hh.GetMinimumBin()))
        hmax = max(hmax, hh.GetBinContent(hh.GetMaximumBin()))

    hists.append(hhs)
    legends.append(legend)

### Set Axis Ranges
if hmin - 5*BOT_MARGIN*(hmax-hmin) < 0:
    hmin, hmax = 0, hmax + TOP_MARGIN*hmax
else:
    mult = (1 + BOT_MARGIN + TOP_MARGIN) * (hmax - hmin)
    hmin, hmax = hmin - BOT_MARGIN*mult, hmax + TOP_MARGIN*mult

### Draw
for ihhs, (hhs, lg) in enumerate(zip(hists,legends)):
    canvas.cd(ihhs + 1).SetGrid()
    canvas.cd(ihhs + 1).SetRightMargin(0)
    for hh in hhs:
        hh.SetMinimum(hmin)
        hh.SetMaximum(hmax)
        hh.Draw("same")
    lg.Draw()

canvas.Draw()
```

```python
# Light Yield Per Channel, Module, Source

# +----------+----------+----------+
# | module 0 | module 1 | module 2 |
# +----------+----------+----------+
# |    SPE   |    SPE   |    SPE   |
# |  SODIUM  |  SODIUM  |  SODIUM  |
# |  CESIUM  |  CESIUM  |  CESIUM  |
# |  COBALT  |  COBALT  |  COBALT  |
# +----------+----------+----------+

n_cols, n_rows = len(module_serial_nums), len(module_sources)
canvas = TCanvas("","",n_cols*TPAD_WIDTH,n_rows*TPAD_HEIGHT)
canvas.Divide(n_cols,n_rows)
# print(n_cols, n_rows)

source_ly = {s: [] for s in module_sources}
for im, module_sn in enumerate(module_serial_nums):
    for js, source in enumerate(module_sources):
        canvas.cd(im + n_cols*js + 1).Clear()
        canvas.cd(im + n_cols*js + 1).SetGrid()
        canvas.cd(im + n_cols*js + 1).SetRightMargin(0)

        module_id = f"{module_sn}_{'lyso' if source == 'spe' else source}"
        if module_id in modules:
            tfile = modules[module_id]["tfile"]
            tfile_keys = [k.GetName() for k in tfile.GetListOfKeys()]
            barsL, barsR = [], []

            lyL, lyEL, lyR, lyER = [], [], [], []
            if source == "spe":
                iLY, iLYE = 3, 5
                hmin, hmax = 2, 5
            elif source == "lyso":
                iLY, iLYE = 3, 5
                hmin, hmax = -1e3, 1e3
            else:
                iLY, iLYE = 1, 2
                hmin, hmax = 30, 100

            for bar in BARS:
                if f"{source}_ch{bar}_fit" in tfile_keys:#modules[module_id]["tfile"]:
                    fit = tfile.Get(f"{source}_ch{bar}_fit")
                    lyL.append(fit.GetParameters()[iLY]) # ave spe charge
                    # lyEL.append(fit.GetParErrors()[iLY]) # spe charge fit error
                    lyEL.append(fit.GetParameters()[iLYE]) # spe charge std
                    barsL.append(bar)
                if f"{source}_ch{bar + 16}_fit" in tfile_keys:#modules[module_id]["tfile"]:
                    fit = tfile.Get(f"{source}_ch{bar + 16}_fit")
                    lyR.append(fit.GetParameters()[iLY]) # ave spe charge
                    # lyER.append(fit.GetParErrors()[iLY]) # spe charge fit error
                    lyER.append(fit.GetParameters()[iLYE]) # spe charge std
                    barsR.append(bar)
            barsL, barsR = np.array(barsL, dtype=float), np.array(barsR, dtype=float)
            lyL, lyEL = np.array(lyL, dtype=float), np.array(lyEL, dtype=float)
            lyR, lyER = np.array(lyR, dtype=float), np.array(lyER, dtype=float)
            lyEL, lyER = np.abs(lyEL), np.abs(lyER)
            # print(lyEL, lyER)
            if "sodium" in module_id:
                print(module_id, (np.mean(lyL), np.mean(lyEL)), (np.mean(lyR), np.mean(lyER)))
            legend = TLegend(0.8, 0.75, 0.97, 0.87)

            if len(lyL):
                grL = TGraphErrors(len(lyL),barsL,lyL,0*barsL,lyEL)
                legend.AddEntry(grL, "Left Channels", "L")
                grL.SetTitle(f"{source.upper()} LY/bar for Module {module_sn}")
                grL.SetLineColor(STD_COLOR_LIST[0])
                grL.GetXaxis().SetTitle("Bar")
                grL.GetYaxis().SetTitle("Charge (pC)")
                grL.SetLineWidth(4)
                grL.SetFillColorAlpha(STD_COLOR_LIST[0], 0.3)
                grL.SetMinimum(hmin)
                grL.SetMaximum(hmax)
                grL.Draw("A PCE4")
                gc.append(grL)

            if len(lyR):
                grR = TGraphErrors(len(lyR),barsR,lyR,0*barsR,lyER)
                legend.AddEntry(grR, "Right Channels", "L")
                grR.SetTitle(f"{source.upper()} LY/bar for Module {module_sn}")
                grR.GetXaxis().SetTitle("Bar")
                grR.GetYaxis().SetTitle("Charge (pC)")
                grR.SetLineColor(STD_COLOR_LIST[1])
                grR.SetLineWidth(4)
                grR.SetFillColorAlpha(STD_COLOR_LIST[1], 0.3)
                grR.SetMinimum(hmin)
                grR.SetMaximum(hmax)
                grR.Draw(("" if len(lyL) else "A") + " PCE4")
                gc.append(grR)
            if len(lyL)+len(lyR):
                legend.Draw()
                gc.append(legend)

canvas.Draw()
```
