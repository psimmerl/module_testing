```python
# Load Data
SOURCE_ENERGY_FIT_PAR_IDXS = {"spe": (1, 3,5), "lyso": (1, 3,5), "sodium": (511,0,1), "cesium": (662,0,1), "cobalt": (122,0,1)}

modules, module_serial_nums = {}, []
for fname in os.listdir(DATA_DIR):
    if "root" not in fname:
        continue

    stats = fname.split("_")
    if stats[1] not in module_serial_nums:
        module_serial_nums.append(stats[1])

    source = [s for s in SOURCES if s in stats[4]][0]
    module_id =  f"{stats[1]}_{source}"

    modules[module_id] = {
        "ov" : float(stats[2][3:]),
        "source" : source,
        "n_spe" : int(stats[3][4:]),
        "n_source" : int(stats[4][len(source)+1:-5]),
        "tfile" : TFile(F"{DATA_DIR}/{fname}"),
        # "tfile_keys": [k.GetName() for k in TFile(F"{DATA_DIR}/{fname}").GetListOfKeys()],
        "spe_charge": np.empty((32,2), dtype=float), # channels x (val, err)
        "source_charge": np.empty((32,2), dtype=float), # channels x (val, err)
    }
    modules[module_id]["tfile_keys"] = [k.GetName() for k in modules[module_id]["tfile"].GetListOfKeys()]

    ## Extract Channel SPE & Source Charges
    for s in ("spe", "source"):
        _s = source if s == "source" else s
        eng, iAve, iErr = SOURCE_ENERGY_FIT_PAR_IDXS[_s]

        for ch in CHANNELS:
            k = f"{_s}_ch{ch}_fit"
            if k not in modules[module_id]["tfile_keys"]:
                continue

            tfit = modules[module_id]["tfile"].Get(k)
            if iAve == iErr:
                modules[module_id][f"{s}_charge"][ch] = [eng*tfit.GetParameter(iAve), np.abs(tfit.GetParError(iAve))]
            else:
                modules[module_id][f"{s}_charge"][ch] = [eng*tfit.GetParameter(iAve), np.abs(tfit.GetParameter(iErr))]

module_sources = ["spe"] + [ s for s in SOURCES if s in [m["source"] for mid,m in modules.items()] ]
```

```python
# Plot SPE for each bar

## Settings
serial_num = module_serial_nums[0]
source = ("spe", "lyso", "sodium", "cesium", "cobalt")[0]
module_id = f"{serial_num}_{'lyso' if source == 'spe' else source}"

for serial_num in module_serial_nums:
    for source in SOURCES:
        module_id = f"{serial_num}_{'lyso' if source == 'spe' else source}"
        if module_id not in modules:
            continue

        print(f"{serial_num=}")
        print(f"{module_id=}")
        print(f"{source=}")

        ## Plots
        n_rows, n_cols = max(floor(np.sqrt(len(BARS))),3), max(ceil(np.sqrt(len(BARS))),3)
        canvas = TCanvas("","",TPAD_WIDTH*n_cols,TPAD_HEIGHT*n_rows)
        canvas.Divide(n_cols, n_rows)
        tlegend_coords = (0.8, 0.75, 0.97, 0.87)

        module = modules[module_id]
        tfile = module["tfile"]

        ### Create Histos
        hists, legends, hmin, hmax = [], [], 1e99, -1
        for bar in BARS:
            chs = [bar, bar+16] # [left_ch, right_ch]
            nns = [f"Ch {ch}" for ch in chs]
            hhs = [tfile.Get(f"{source}_ch{ch}") for ch in chs]
            tts = [f"Bar {bar} {source} Response"] * len(chs)
            lws = [10] * len(chs)
            lcs = [rt.kAzure, rt.kOrange]

            legend = TLegend(*tlegend_coords)
            for ihh, hh in enumerate(hhs):
                if len(hh.GetListOfFunctions()):
                    hh.GetListOfFunctions()[0].SetLineWidth(int(lws[ihh]/2))
                hh.SetTitle(tts[ihh])
                hh.SetName(nns[ihh])
                hh.SetLineColor(lcs[ihh])
                hh.SetLineWidth(lws[ihh])
                legend.AddEntry(hh, f"{nns[ihh]}", "l")
                hmin = min(hmin, hh.GetBinContent(hh.GetMinimumBin()))
                hmax = max(hmax, hh.GetBinContent(hh.GetMaximumBin()))

            hhs.sort(key=lambda h: h.GetXaxis().GetXmin())
            hists.append(hhs)
            legends.append(legend)

        ### Set Axis Ranges
        if hmin - 5*BOT_MARGIN*(hmax-hmin) < 0:
            hmin, hmax = 0, hmax + TOP_MARGIN*hmax
        else:
            mult = (1 + BOT_MARGIN + TOP_MARGIN) * (hmax - hmin)
            hmin, hmax = hmin - BOT_MARGIN*mult, hmax + TOP_MARGIN*mult

        ### Draw
        for ihhs, (hhs, lg) in enumerate(zip(hists,legends)):
            canvas.cd(ihhs + 1).SetGrid()
            canvas.cd(ihhs + 1).SetRightMargin(0)
            for hh in hhs:
                hh.SetMinimum(hmin)
                hh.SetMaximum(hmax)
                hh.Draw("same")
            lg.Draw()

        canvas.Draw()
        canvas.Print(f"{PLOT_DIR}/module_{serial_num}_{source}_spectra.png")
```

```python
# Light Yield Per Channel, Module, Source

# +----------+----------+----------+
# | module 0 | module 1 | module 2 |
# +----------+----------+----------+
# |    SPE   |    SPE   |    SPE   |
# |  SODIUM  |  SODIUM  |  SODIUM  |
# |  CESIUM  |  CESIUM  |  CESIUM  |
# |  COBALT  |  COBALT  |  COBALT  |
# +----------+----------+----------+
NORM_BY = "measured_charge"
# NORM_BY = "side_ave"
# NORM_BY = "spe_ch"
# NORM_BY = "spe_side_ave"

for NORM_BY in ("measured_charge", "side_ave", "spe_ch", "spe_side_ave"):
    _module_sources = [s for s in module_sources if not("spe" in NORM_BY and "spe" in s)]
    print(_module_sources, NORM_BY)
    
    ################################
    ################################
    ################################
    
    n_cols, n_rows = len(module_serial_nums), len(_module_sources)
    canvas = TCanvas("","",n_cols*TPAD_WIDTH,n_rows*TPAD_HEIGHT)
    canvas.Divide(n_cols,n_rows)
    # print(n_cols, n_rows)
    
    source_ly = {s: [] for s in module_sources}
    for im, module_sn in enumerate(module_serial_nums):
        for js, source in enumerate(_module_sources):
            canvas.cd(im + n_cols*js + 1).Clear()
            canvas.cd(im + n_cols*js + 1).SetGrid()
            canvas.cd(im + n_cols*js + 1).SetRightMargin(0)
    
            module_id = f"{module_sn}_{'lyso' if source == 'spe' else source}"
            if module_id in modules:
                tfile = modules[module_id]["tfile"]
    
                ly_spe = modules[module_id]["spe_charge"]
                ly_src = modules[module_id]["source_charge"]
                ly = ly_spe if source == "spe" else ly_src
    
                ### Set Axis Ranges
                hmin, hmax = {"spe": (0, 5),
                              "lyso": (-1e3, 1e3),
                              "sodium": (0, 1500),
                              "cesium": (0, 2000),
                              "cobalt": (0, 400)}[source]
                # hmin, hmax = np.min(ly[:,0]-ly[:,1]), np.max(ly[:,0]+ly[:,1])
                # print(module_id, igr, source, hmin, hmax)
                # _BOT_MARGIN, _TOP_MARGIN = 0.1, 0.1
                # if hmin - 5*_BOT_MARGIN*(hmax-hmin) < 0:
                #     hmin, hmax = 0, hmax + _TOP_MARGIN*hmax
                # else:
                #     mult = (1 + _BOT_MARGIN + _TOP_MARGIN) * (hmax - hmin)
                #     hmin, hmax = hmin - _BOT_MARGIN*mult, hmax + _TOP_MARGIN*mult
                # print(module_id, igr, source, hmin, hmax)
                # hmin=0
    
                ### Build Graphs
                ly_spes, ly_speEs = [ly_spe[:16,0], ly_spe[16:,0]], [ly_spe[:16,1], ly_spe[16:,1]]
                lys, lyEs = [ly[:16,0], ly[16:,0]], [ly[:16,1], ly[16:,1]]
                bars = [BARS] * len(lys)
                tt = f"{source} LY/bar for Module {module_sn}"
                xl, yl = "bar", "Charge (pC)"
    
                if NORM_BY == "side_ave":
                    tt = f"{source} LY/bar for Module {module_sn}, Normalized by Side's Ave Charge"
                    yl = "Channel Charge / Side's Average Charge"
                    lyEs = [np.sqrt(np.sum(np.divide(lyE,ly,where=ly!=0,out=np.zeros_like(lyE))**2)
                            + np.sum(lyE**2)/np.sum(ly)**2) for ly, lyE in zip(lys, lyEs)]
                    lys = [ly/np.mean(ly) for ly in lys]
                    lyEs = [ly * lyE for ly, lyE in zip(lys, lyEs)]
                    hmin, hmax = {"spe": (0, 2),
                      "lyso": (-1e3, 1e3),
                      "sodium": (0, 2),
                      "cesium": (0, 2),
                      "cobalt": (0, 2)}[source]
                elif NORM_BY == "spe_ch":
                    tt = f"{source} LY/bar for Module {module_sn}, Normalized by Channel's SPE Charge"
                    yl = "Channel Charge / SPE Charge"
                    lyEs = [np.sqrt(
                        np.divide(lyE,ly,where=ly!=0,out=np.zeros_like(lyE))**2
                        + np.divide(lySpeE,lySpe,where=lySpe!=0,out=np.zeros_like(lySpeE))**2)
                            for ly, lyE, lySpe, lySpeE in zip(lys, lyEs, ly_spes, ly_speEs)]
                    lys = [np.divide(ly,lySpe,where=lySpe!=0,out=np.zeros_like(ly)) for ly, lySpe in zip(lys, ly_spes)]
                    lyEs = [ly * lyE for ly, lyE in zip(lys, lyEs)]
                    hmin, hmax = {"spe": (0, 5),
                      "lyso": (-1e3, 1e3),
                      "sodium": (0, 450),
                      "cesium": (0, 450),
                      "cobalt": (0, 450)}[source]
                elif NORM_BY == "spe_side_ave":
                    tt = f"{source} LY/bar for Module {module_sn}, Normalized by Side's SPE Charge"
                    yl = "Channel Charge / Side's Average SPE Charge"
                    lyEs = [np.sqrt(np.sum(np.divide(lyE,ly,where=ly!=0,out=np.zeros_like(lyE))**2)
                            + np.sum(lySpeE**2)/np.sum(lySpe)**2) for ly, lyE, lySpe, lySpeE in zip(lys, lyEs, ly_spes, ly_speEs)]
                    lys = [ly/np.mean(lySpe) for ly, lySpe in zip(lys, ly_spes)]
                    lyEs = [ly * lyE for ly, lyE in zip(lys, lyEs)]
                    hmin, hmax = {"spe": (0, 5),
                      "lyso": (-1e3, 1e3),
                      "sodium": (0, 450),
                      "cesium": (0, 450),
                      "cobalt": (0, 450)}[source]
    
                lls = ["Left (A)", "Right (B)"]
                lcs = [rt.kAzure, rt.kOrange]
                lws = [6] * len(lls)
    
                legend = TLegend(0.75, 0.13, 0.97, 0.28)
                # legend = TLegend(0.8, 0.75, 0.97, 0.87)
                for igr in range(len(nns)):
                    # if "sodium" in module_id:
                    _ly = np.array(lys[igr], dtype=float)
                    _lyE = np.array(lyEs[igr], dtype=float)
                    _bars = np.array(bars[igr], dtype=float)
                    # print(module_id, source, igr, _bars, _ly, _lyE)#np.mean(_ly), np.mean(_lyE))
    
                    gr = TGraphErrors(len(_ly),_bars,_ly,0*_bars,_lyE)
                    # gr = TGraphErrors(len(_ly),_bars,_bars,_bars,_bars)#,_ly,0*_bars,_lyE)
                    legend.AddEntry(gr, lls[igr], "L")
                    gr.SetTitle(tt)
                    gr.GetXaxis().SetTitle(xl)
                    gr.GetYaxis().SetTitle(yl)
    
                    gr.SetMarkerStyle(rt.kFullCircle)
                    gr.SetMarkerSize(lws[igr]//3)
                    gr.SetMarkerColor(lcs[igr])
                    gr.SetLineWidth(lws[igr])
                    gr.SetLineColor(lcs[igr])
                    gr.SetFillColorAlpha(lcs[igr], 0.3)
    
                    gr.SetMinimum(hmin)
                    gr.SetMaximum(hmax)
                    gr.GetXaxis().SetLimits(-0.5, 15.5)
                    gr.Draw(("" if igr else "A") + " PCE4")
    
                    gc.append(gr)
                legend.Draw()
                gc.append(legend)
    
    canvas.Draw()
    canvas.Print(f"{PLOT_DIR}/ly_{NORM_BY}.png")
```
